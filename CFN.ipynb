```python
#check for the coloumns which have outliers 
numerical_cols = df.select_dtypes(include=np.number).columns
display(numerical_cols)
```

```python
#calculating the outliers
print("--- Outlier Report ---")

# 2. Loop through each numerical column
for col in numerical_cols:
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    
    # Calculate the Interquartile Range (IQR)
    IQR = Q3 - Q1
    
    # Define the outlier boundaries 
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # 3. Find the data points that are considered outliers
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    
    # 4. Print the results
    if not outliers.empty:
        print(f"Column '{col}': Found {len(outliers)} outliers.")
    else:
        print(f"Column '{col}': No outliers found.")

```

```python
#columns with outliers > 0
columns_to_plot = [
    'fuel_consumption_rate',
    'iot_temperature',
    'disruption_likelihood_score'
]

sns.set_style("whitegrid")

# Create a figure to hold the plots
plt.figure(figsize=(18, 7))

# Generate a boxplot for each column
for i, col in enumerate(columns_to_plot, 1):
    plt.subplot(1, 3, i) # Create a 1x3 grid of plots
    sns.boxplot(y=df[col]) # Create a vertical boxplot
    plt.title(f'Box Plot of {col}', fontsize=14)
    plt.ylabel('Value')

plt.tight_layout()
plt.show()
```

```python
#finding the corealtion using a heatmap
numerical_df = df.select_dtypes(include=['float64', 'int64'])

# 2. Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

# 3. Set up and draw the heatmap
plt.figure(figsize=(20, 16))
sns.heatmap(correlation_matrix, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features', fontsize=16)

# 4. Show the plot
plt.show()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])

# --- Feature Engineering ---
df['hour_of_day'] = df['timestamp'].dt.hour
df['day_of_week'] = df['timestamp'].dt.day_name()
df['day_of_month'] = df['timestamp'].dt.day
df['month'] = df['timestamp'].dt.month
df['is_weekend'] = df['timestamp'].dt.weekday >= 5 # Monday=0, Sunday=6

# Define peak hours (e.g., 7-9 AM and 4-6 PM)
peak_hours = [7, 8, 9, 16, 17, 18]
df['is_peak_hour'] = df['hour_of_day'].isin(peak_hours)

# Display the new features for the first 5 rows
print("DataFrame with new time-based features:")
print(df[['timestamp', 'hour_of_day', 'day_of_week', 'is_weekend', 'is_peak_hour']].head())
```

```python
# 2. Convert 'timestamp' to a datetime format
df['timestamp'] = pd.to_datetime(df['timestamp'])

# 3. Separate into 'date' and 'time' columns
df['date'] = df['timestamp'].dt.date
df['time'] = df['timestamp'].dt.time

# --- Display the result ---
print(df[['timestamp', 'date', 'time']].head())
```

```python
# --- Visualization of Operational Patterns ---
fig, axes = plt.subplots(1, 2, figsize=(20, 7))

# Weekly Pattern
day_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
sns.countplot(data=df, x='day_of_week', ax=axes[0], order=day_order, palette='viridis')
axes[0].set_title('Total Truck Activity by Day of the Week', fontsize=14)
axes[0].set_xlabel('Day of the Week', fontsize=12)
axes[0].set_ylabel('Total Telematics Readings', fontsize=12)

# Daily Pattern
sns.lineplot(data=df, x='hour_of_day', y='traffic_congestion_level', ax=axes[1], color='crimson')
axes[1].set_title('Average Traffic Congestion Throughout the Day', fontsize=14)
axes[1].set_xlabel('Hour of the Day (24-Hour Clock)', fontsize=12)
axes[1].set_ylabel('Average Traffic Congestion Score', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# --- Geospatial Visualization ---
plt.figure(figsize=(12, 10))
# Sample data to avoid overplotting and improve performance
df_sample = df.sample(n=5000, random_state=1)
sns.scatterplot(data=df_sample, x='vehicle_gps_longitude', y='vehicle_gps_latitude', hue='route_risk_level', s=10, alpha=0.6, palette='magma')
plt.title('Geographic Hotspots of Truck Routes by Risk Level', fontsize=16)
plt.xlabel('Longitude', fontsize=12)
plt.ylabel('Latitude', fontsize=12)
plt.legend(title='Route Risk Level')
plt.grid(False)
plt.show()
```
